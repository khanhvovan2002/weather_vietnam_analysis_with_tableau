# This work involves analysis and visualization with all regions Viet nam  climate. Making API requests to OpenWeatherMap's historical weather data API 

```python
import requests, json
import pandas as pd
import numpy as np
import time
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import ast
import datetime as dt
```

We need to filter all the samples from country VietNam from this json file.(The json contain Geographical IDs,for which we  make request api to query the data)

```python
df = pd.read_json("city_list.json")
df
```

```python
df = df.loc[df['country'] == 'VN']
```

```python
df.name.unique()
```

```python
data_grouped = pd.DataFrame()
```

```python
df
```

  Initilize Time range for collecting data

```python
now = int(time.time())
lw = now - 604800
id_ = df.id.unique()
```

This code execute querying data by api with timerange and _id region encrypted and merge the data recieved with the prefiltered dataframe VietNam

```python
id_ = df.id.unique()
for id__ in id_:
    coord_ = pd.DataFrame(df.loc[df['id'] == id__]['coord'])
    province_ = pd.DataFrame(df.loc[df['id'] == id__]['name'])
    coord_data = coord_.reset_index()
    province_data = province_.reset_index()
    data = requests.get("https://history.openweathermap.org/data/2.5/history/city?id="+str(id__)+"&type=hour&start="+str(lw)+"&end="+str(now)+"&units=metric&appid=36de67b0831e2d23f5e496827c24592f")
    data = json.loads(data.content)
    data = pd.DataFrame(data['list'])
    data['id'] = id__
    data.loc[:,'coord'] = np.repeat([coord_data['coord'][0]],data.shape[0])
    data['province'] = np.repeat([province_data['name'][0]],data.shape[0])
    data_grouped = pd.concat([data, data_grouped])
```

```python
data_grouped
```

```python
data_grouped.to_csv('data_final.csv')
```

```python
data_new = pd.read_csv('data_final.csv')
      
```

So this is the merged dataframe we just made, some columns have diction format, the time feature is  in unix time format. We need to reorganize the dataframe to an more simple format. 

```python
data_new
```

```python
data_new.province.unique()
```

Notice the type of the weather main coord columns , they are string types, so we convert it into dict type , for easy processing

```python
data_new['weather']=data_new.weather.map(lambda value:  ast.literal_eval(value))

data_new['main'] = data_new.main.map(lambda value:  ast.literal_eval(value))
data_new['coord']=data_new.coord.map(lambda value:  ast.literal_eval(value))
```

```python
data_new['wind']=data_new.wind.map(lambda value:  ast.literal_eval(value))
```

```python
data_new['clouds']=data_new.clouds.map(lambda value:  ast.literal_eval(value))
```

```python
data_new = data_new.explode('weather')
```

```python
data_new
```

We flatten the dict columns and convert the feature time into an good format

```python
main = pd.json_normalize(data_new['main'], max_level = 1)
wind = pd.json_normalize(data_new['wind'], max_level =1)
clouds = pd.json_normalize(data_new['clouds'], max_level =1)
weather = pd.json_normalize(data_new['weather'], max_level =1)
coord = pd.json_normalize(data_new['coord'], max_level= 1)
```

```python
data_new.drop(columns = ['main','wind','clouds','weather','coord'],inplace = True)
```

```python

data_new['dt'] = pd.to_datetime(data_new['dt'], unit='s')
data_new = data_new.set_index('dt')
data_new = data_new.tz_localize('UTC')
data_new = data_new.tz_convert('Asia/Ho_Chi_Minh')
data_new.reset_index(inplace = True)
data_new['dt']=data_new['dt'].dt.tz_localize(None)
data_new = pd.concat([data_new, main, wind, clouds, weather, coord], axis =1 )
data_new = data_new.set_index('dt')
data_new
```

```python
data_new.drop(columns = ['rain', 'Unnamed: 0', 'icon'], inplace = True)
```
